### Interaktives Rechenwerkzeug: Wahrheitstabelle
Geben Sie einen aussagenlogischen Ausdruck ein. Verwenden Sie die folgenden Symbole:

    UND: &
    ODER: |
    NICHT: ~
    Implikation: >>
    Äquivalenz: == Beispiel: (p & q) >> r

```python
import sympy as sp
import itertools

def erstelle_wahrheitstabelle(ausdruck_str):
    # Definition der möglichen Variablen
    all_vars = 'pqrstuvw'
    syms = {char: sp.symbols(char) for char in all_vars}
    
    try:
        # Ausdruck in SymPy-Logik umwandeln
        expr = sp.sympify(ausdruck_str, locals=syms)
        
        # Finden, welche Variablen tatsächlich im Ausdruck vorkommen
        used_vars = sorted(list(expr.free_symbols), key=lambda s: s.name)
        
        # Header für die Tabelle erstellen
        header = " | ".join([str(v) for v in used_vars]) + " | Ergebnis"
        print(header)
        print("-" * len(header))
        
        # Alle möglichen Wahrheitsbelegungen generieren (True/False)
        # product([True, False], repeat=n) erzeugt alle Kombinationen
        combinations = list(itertools.product([True, False], repeat=len(used_vars)))
        
        all_true = True
        
        for combo in combinations:
            # Belegung erstellen: {p: True, q: False, ...}
            subs_dict = dict(zip(used_vars, combo))
            
            # Ausdruck auswerten
            result = expr.subs(subs_dict)
            
            # Formatierung für die Ausgabe (T für True, F für False)
            row_vals = [ "T" if val else "F" for val in combo ]
            res_val = "T" if result else "F"
            
            print(f"{' | '.join(row_vals)} | {res_val}")
            
            if not result:
                all_true = False
        
        print("-" * len(header))
        if all_true:
            print("Ergebnis: Der Ausdruck ist eine TAUTOLOGIE.")
        else:
            print("Ergebnis: Der Ausdruck ist KEINE Tautologie.")
            
    except Exception as e:
        print(f"Fehler: Bitte prüfen Sie die Syntax. {e}")
```

```python
# --- Eingabebereich ---
# Beispiel: "(p & q) >> r" oder "(p >> q) == (~q >> ~p)"
mein_ausdruck = "(p & q) >> r"

erstelle_wahrheitstabelle(mein_ausdruck)
```
